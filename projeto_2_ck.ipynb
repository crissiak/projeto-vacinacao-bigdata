```python
# Projeto 2 - Análise de Vacinação COVID-19 (Camada Gold)
# FUNDAMENTOS E PLATAFORMA DE BIG DATA

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Leitura da camada Gold (arquivo CSV)
df = pd.read_csv("vacinacao_gold.csv")

# Visualizar os primeiros registros
df.head()
```

#Total de Doses por Estado

```python
# Agrupando total de doses por estado
total_por_estado = df.groupby('estado')['total_doses'].sum().reset_index()

# Gráfico de barras
plt.figure(figsize=(8, 5))
sns.barplot(data=total_por_estado, x='estado', y='total_doses', palette='Blues_d')
plt.title("Total de Doses Aplicadas por Estado")
plt.xlabel("Estado")
plt.ylabel("Total de Doses")
plt.tight_layout()
plt.show()
```

#Comparação Geral: 1ª vs 2ª Dose

```python
# Agrupamento por tipo de dose
comparacao_dose = df.groupby('dose')['total_doses'].sum().reset_index()

# Gráfico comparativo
plt.figure(figsize=(6, 5))
sns.barplot(data=comparacao_dose, x='dose', y='total_doses', palette='Set2')
plt.title("Comparação Geral: 1ª vs 2ª Dose")
plt.xlabel("Tipo de Dose")
plt.ylabel("Total de Doses")
plt.tight_layout()
plt.show()
```

#Heatmap de Doses por Estado e Tipo

```python
# Tabela cruzada de doses por estado e tipo
tabela = df.pivot(index='estado', columns='dose', values='total_doses').fillna(0)

# Heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(tabela, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Doses Aplicadas por Estado e Tipo")
plt.tight_layout()
plt.show()
```
